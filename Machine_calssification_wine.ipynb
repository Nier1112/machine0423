```python
from sklearn.datasets import load_wine
wine = load_wine()
wine
# data, 要處理的數據, target:分類標籤, target_names:分成幾類, 
```

```python
#整理成 DataFrame
import pandas as pd
df = pd.DataFrame(wine["data"], columns=wine["feature_names"])
df["target"] = wine["target"]
df
```

```python
#畫出關係圖
df.corr()
```

```python
#畫出熱度圖
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20, 15))
sns.heatmap(df.astype("float").corr(), cmap="OrRd", annot=True)
```

```python
# 切割資料: 90%訓練, 10%測試
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(["target"], axis=1), df["target"], test_size=0.1)
```

```python
#訓練模型
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=3)
clf.fit(x_train, y_train)
```

```python
#畫決策樹
from sklearn.tree import export_graphviz
import graphviz
g = export_graphviz(clf, out_file=None, feature_names=wine["feature_names"], class_names=wine["target_names"], filled=True, special_characters=True)
graph = graphviz.Source(g)
graph
```

```python
#預測, 把 x 測試資料做預測
pre = clf.predict(x_test)
print("預測:", list(pre))
print("實際:", list(y_test))
```

```python
#算出機率
from sklearn.metrics import accuracy_score
print("預測成功機率", accuracy_score(y_test, pre))
```
