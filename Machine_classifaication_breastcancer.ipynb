```python
# breast_cancer : https://scikit-learn.org/stable/datasets/index.html
from sklearn.datasets import load_breast_cancer
bc = load_breast_cancer()
bc
```

```python
#整理成data
import pandas as pd
df = pd.DataFrame(bc["data"], columns=bc["feature_names"])
df["target"] = bc["target"]
df
```

```python
#關係圖
df.corr()
```

```python
#熱度圖
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(25, 20))
sns.heatmap(df.astype("float").corr(), cmap="OrRd", annot=True)
```

```python
#切割資料
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(["target"], axis=1), df["target"], test_size=0.1)
```

```python
#製作決策樹
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=3)
clf.fit(x_train, y_train)
```

```python
#畫出決策樹
from sklearn.tree import export_graphviz
import graphviz
g = export_graphviz(clf, feature_names=bc["feature_names"], class_names=bc["target_names"], filled=True, special_characters=True)
graphviz.Source(g)
```

```python
#predict 預測
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test)
print("預測:", list(pre))
print("真實:", list(y_test))
print("正確率:", accuracy_score(y_test, pre))
```
